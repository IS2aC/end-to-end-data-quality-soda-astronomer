```python
import pandas as pd 
from datetime import datetime
```

## YELLOW TAXI ...

```python
data_yellow =  pd.read_parquet("google_sheet/yellow_tripdata_2024-01.parquet", engine="pyarrow")
data_yellow = data_yellow.sort_values(by='tpep_pickup_datetime', ascending =  True)
data_yellow['date'] = data_yellow['tpep_pickup_datetime'].dt.date
data_yellow =  data_yellow[data_yellow['tpep_pickup_datetime'] >='2024-01-01']
data_yellow.shape
```

```python
# visualiez schema
data_yellow.dtypes
```

```python
data_yellow['date'].unique()
```

```python
list_row =  []
for i in data_yellow['date'].unique():
    dt = data_yellow[data_yellow['date'] == i]
    dt =  dt.tail(3000)
    list_row.append(dt)
```

```python
df_yellow =  pd.concat(list_row)
df_yellow
```

```python
df_yellow['taxi'] =  'YELLOW'
```

```python
df_yellow.to_excel('DATASET_NYC_YELLOW.xlsx', index  =  False)
```

## Data Catalog


| **Colonne**                          | **Type de données**           | **Description**                                                                 |
|---------------------------------|-------------------------------|---------------------------------------------------------------------------------|
| `VendorID`                      | int32                         | Identifiant du fournisseur de taxi (1 = Creative Mobile Technologies, 2 = Verifone).|
| `tpep_pickup_datetime`          | datetime64[us]                | Date et heure de la prise en charge du passager.                                |
| `tpep_dropoff_datetime`         | datetime64[us]                | Date et heure de la dépose du passager.                                         |
| `passenger_count`               | float64                       | Nombre de passagers dans le taxi.                                               |
| `trip_distance`                 | float64                       | Distance totale du trajet (en miles).                                           |
| `RatecodeID`                    | float64                       | Code du tarif appliqué (par exemple, tarif standard ou tarif aéroport).         |
| `store_and_fwd_flag`            | object                         | Indicateur (Y/N) si le trajet a été enregistré en mode "store-and-forward" pendant une interruption de réseau.|
| `PULocationID`                  | int32                         | ID de la zone de prise en charge (zone géographique de départ).                 |
| `DOLocationID`                  | int32                         | ID de la zone de dépose (zone géographique d'arrivée).                          |
| `payment_type`                  | int64                         | Mode de paiement utilisé (par exemple, carte de crédit, espèces, etc.).         |
| `fare_amount`                   | float64                       | Montant de la course (tarif de base).                                           |
| `extra`                         | float64                       | Montant supplémentaire appliqué (par exemple, surcharge pour certains services).|
| `mta_tax`                       | float64                       | Taxe MTA (Metropolitan Transportation Authority) appliquée à la course.        |
| `tip_amount`                    | float64                       | Montant du pourboire donné au chauffeur.                                        |
| `tolls_amount`                  | float64                       | Montant des péages payés pendant le trajet.                                     |
| `improvement_surcharge`         | float64                       | Surcharge pour l'amélioration des infrastructures.                              |
| `total_amount`                  | float64                       | Montant total de la course, incluant tous les frais et taxes.                   |
| `congestion_surcharge`          | float64                       | Surcharge pour la congestion dans certaines zones de la ville.                 |
| `Airport_fee`                   | float64                       | Frais d'aéroport, généralement appliqué pour les trajets à destination ou en provenance de l'aéroport. |
